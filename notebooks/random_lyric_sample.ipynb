```python
import pandas as pd
import numpy as np
import random


random.seed(119)


DATA_DIR = "/content/drive/MyDrive/W266 Project/w266-finalproj/data"
```

```python
npvalid_labels = ['Folk','Jazz','Metal','Pop','R&B','Rap','Rock','Soul']
```

```python
original = pd.read_csv(f"{DATA_DIR}/16_tokens_seeds.csv", index_col=0)
original
```

```python
gen_data = pd.DataFrame(columns=['human', 'baseline', 'attention'])
genres = ['Folk','Jazz','Metal','Pop','RB','Rap','Rock','Soul']
for genre in genres:
  baseline = pd.read_csv(f"{DATA_DIR}/baseline_{genre.lower()}_generated_lyrics.csv", index_col=0)

  if genre == 'RB':
    genre = 'R&B'

  attention = pd.read_csv(f"{DATA_DIR}/lstm_attention_{genre.lower()}_generated_lyrics.csv", index_col=0)
  humangen = original[original['genre']==genre]

  human_lyric = humangen.sample(n=3, random_state=119, replace=False)
  baseline_lyric = baseline.sample(n=3, random_state=312, replace=False)
  attention_lyric = attention.sample(n=3, random_state=412, replace=False)
  human_lyric['gen_type'] = 'human'
  baseline_lyric['gen_type'] = 'baseline'
  attention_lyric['gen_type'] = 'attention'

  human_lyric['genre'] = genre
  baseline_lyric['genre'] = genre
  attention_lyric['genre'] = genre

  human_lyric = human_lyric[['gen_type', 'genre', 'lyric']]
  baseline_lyric = baseline_lyric[['gen_type', 'genre', 'generated_lyric']]
  attention_lyric = attention_lyric[['gen_type', 'genre', genre]]

  human_lyric = human_lyric.rename(columns={'lyric': 'generated_lyric'})
  
  attention_lyric = attention_lyric.rename(columns={genre: 'generated_lyric'})

  df = pd.concat([human_lyric, baseline_lyric, attention_lyric], axis = 0).reset_index(drop=True)
  df = df.sample(frac=1, replace=False)
  df.to_csv(f"{DATA_DIR}/04_sample/{genre}_lyricgen.csv", index=False)
```
